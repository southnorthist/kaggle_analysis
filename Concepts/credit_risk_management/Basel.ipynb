## 1. Capital
### 1.1 Definition
For Basel Framework, capital here means Regulatory Capital. It contains three categories to classify banks' financial instruments:
1. Common Equity Tier 1
    * Common shares issued by the bank that meet the criteria for classification as common shares for regulatory purposes
    * Stock surplus (share premium) resulting from the issue of instruments included Common Equity Tier 1
    * Retained earnings
    * Accumulated other comprehensive income and other disclosed reserves
    * Common shares issued by consolidated subsidiaries of the bank and held by third parties that meet the criteria for inclusion in Common Equity Tier 1 capital
    * Regulatory adjustments applied in the calculation of Common Equity Tier 1
2. Additional Tier 1
    * instruments issued by the bank that meeting the criteria for inclusion in Additional Tier 1 capital (and are not included in Common Equity Tier 1)
    * stock surplus (share premium) resulting from the issue of instruments included in Additional Tier 1 capital;
    * instruments issued by consolidated subsidiaries of the bank and held by third parties that meet the criteria for inclusion in Additional Tier 1 capital and are not included in Common Equity Tier 1 capital.
    
3. Tier 2 Capital  
    * instruments issued by the bank that meet the criteria for inclusion in Tier 2 capital (and are not included in Tier 1 capital);
    * stock surplus (share premium) resulting from the issue of instruments included in Tier 2 capital;
    * instruments issued by consolidated subsidiaries of the bank and held by third parties that meet the criteria for inclusion in Tier 2 capital and are not included in Tier 1 capital.
    * certain loan-loss provisions
    * regulatory adjustments applied in the calculation of Tier 2 capital

The total Regulatory Capital = Common Equity Tier 1 Capital + Addtional Tier 1 Capital + Tier 2 Capital - Regulatory Adjustment.   
Tier 1 Capital = Common Equity Tier 1 capital + Additional Tier 1 capital - Regulatory Adjustment.

### 1.2 Minimum Requirement for Risk-based Capital
1. Common Equity Tier 1 must be at least 4.5% of risk-weighted assets (RWA).
2. Tier 1 capital must be at least 6% of RWA.
3. Total capital must be at least 8% of RWA.

## 2. Risk Weighted Assets
### 2.1 Definition
Risk-weighted assets are used to determine the minimum amount of capital that must be held by banks and other financial institutions in order to reduce the risk of insolvency. The capital requirement is based on a risk assessment for each type of bank asset.
### 2.2 Calculation
The bank's total RWA is calculated as follows:
$$ \text{RWA}_{\text{Total}} = \text{RWA}_{\text{credit risk}} + \text{RWA}_{\text{market risk}} + \text{RWA}_{\text{operational risk}}$$
Before a bank can calculate RWA for credit risk and RWA for market risk, it must follow the requirements to identify the instruments that are in the trading book. The banking book comprises all instruments that are not in the trading book and all other assets of the bank.
* RWA for <font color='Brick'>**credit risk** </font> is the sum of:
    1. <mark> Credit RWA for banking book exposures except for the following 2-5 expsoures, calculated using: </mark>
        + (a) Standardized Approach
        + (b) Internal Rating Based Approach </mark>
    2. RWA for **counterparty credit risk** arising from banking book exposures and from trading book instruments 
    3. Credit RWA for equity investments in funds that are held in the banking book, calculated using one or more of the approaches:
        + (a) The look-through approach
        + (b) The mandate-based approach
        + (c) The fall-back approach
    4. RWA for **securitisation exposures** held in the banking book, calculated using one or more of the approaches:
        + (a) Securitisation Standardised Approach 
        + (b) Securitisation External Ratings-Based Approach
        + (c) Internal Assessment Approach
        + (d) Securitisation Internal Ratings-Based Approach
        + (e) A risk weight of 1250% in cases where the bank cannot use (a) to (d) above
    5. RWA for **exposures to central counterparties** in the banking book and trading book
    6. RWA for the **risk posed by unsettled transactions and failed trades**, where these transactions are in the banking book or trading book.
* RWA for <font color='Brick'>**market risk** </font> is the sum of:
    1. RWA for market risk for instruments in the trading book and for foreign exchange risk and commodities risk for exposures in the banking book, calculated using:
        + (a) the standardised approach
        + (b) the internal models approach
    2. RWA for credit valuation adjustment (CVA) risk for exposures in the trading book and banking book, calculated using either:
        + (a) the standardised approach for CVA; or
        + (b) the advanced approach for CVA.
* RWA for <font color='Brick'>**operational risk** </font> are calculated using one or more of the following approaches:
    1. the Basic Indicator Approach 
    2. the standardised approaches
    3. the adcanced measurement approach

## 3. Credit Risk RWA
### 3.1 Categorization of expsoures
There are 5 asset clasess and several specialized lending (for corporate exposures) and sub-classes for retail exposures:
* **coporate**: a debt obligation of a corporation, partnership, or proprietorship.  
    * a. project finance (PF)
    * b. object finance (OF)

    * e. high-volatility commercial real estate (HVCRE)  
* **sovereign**: exposures to counterparties treated as sovereigns under the standardised approach.
* **bank**: exposures to banks and those securities firms outlined in the standardised approach.
* **retail**: Exposures to individuals, Residential mortgage loans, Loans extended to small businesses and managed as retail exposures are eligible for retail treatment provided the total exposure of the banking group to a small business borrower is less than â‚¬1 million (CAD 1.25 million).
* **equity**: <mark> TBA </mark>

### 3.2 Standardized Approach
Risk weighted assets (RWA) are designed to address unexpected losses (UL) from exposures. It assigns standardised risk weights to exposures(categorized by Basel). Risk weighted assets are calculated as the product of the standardised risk weights and the exposure amount. Exposures should be risk-weighted net of specific provisions (including partial write-offs).
### 3.3 Internal Rating Based Approach
Subject to certain minimum conditions and disclosure requirements, banks that have received supervisory approval to use the IRB approach may rely on their own internal estimates of risk components in determining the capital requirement for a given exposure. The risk components include measures of the probability of default (PD), loss given default (LGD), the exposure at default (EAD), and effective maturity (M). Depending on which component the bank is allowed to use internally development approach, there are two categories:
* **Foundational IRB Approach**  
Under the foundation approach, banks must provide their own estimates of PD associated with each of their borrower grades, but must use supervisory estimates for the other relevant risk components. The other risk components are LGD, EAD and M.
* **Advanced IRB Approach**  
Under the advanced approach, banks must calculate the effective maturity (M) and provide their own estimates of PD, LGD and EAD.

* **_Note_**: 
    1. Banks that meet the requirement for estimation of PD are able to use **foundation approach** to <font color='steelblue'> **corporate exposures** </font> to derive RWA for all SL classes (exception: HVCRE has special correlation formula).
    2. Banks that meet the requirement for estimation of PD, LGD and EAD are able to use **advanced approach** to <font color='steelblue'> **corporate exposures** </font> to derive RWA for all SL classes (exception: HVCRE has special correlation formula).
    3. Banks that **do not** meet the requirement for estimation of PD under the <font color='steelblue'> **corporate** </font> foundation approach for their SL assets are required to use "supervisory slotting criteria approach" (mapping internal risk grades to five supervisory categories, each of which is associated with a specific risk weight).
    3. For <font color='steelblue'>**retail exposures** </font>, banks must provide their own estimates of PD, LGD and EAD. THere is no distinction between a foundation and advanced approach for this asset class.


### 3.4 Risk Weight Functions
* Probability of default (PD) and loss-given-default (LGD) are measured as decimals. 
* Exposure at default (EAD) is measured as currency (eg euros), except where explicitly noted otherwise.
* N(x) is the standard normal cumulative distribution function
* G(x) is the inverse of the standard normal cumulative distribution function

#### 3.4.1 Risk-weight functions for corporate, sovereign and bank exposures
The formula is as follows:
$$
\begin{aligned}
\text{Correlation} &= R = 0.12 \cdot \frac{1 - \exp^{-50 \cdot PD}}{1 - \exp^{-50}} + 0.24 \cdot \left( 1 - \frac{1 - \exp^{-50 \cdot PD}}{1 - \exp^{-50}} \right) \\
\text{Maturity Adjustment} &= b = \left( 0.11852 - 0.05478 \cdot \ln{PD} \right) ^2\\
\text{Capital Requirement} &= K = \left[ N\left( \frac{1}{\sqrt{1-R}} \cdot G(PD) + \sqrt{\frac{R}{1-R}} \cdot G(0.999) \right) \cdot LGD - PD \cdot LGD \right] \cdot \frac{1 + \left(M - 2.5\right) \cdot b}{1-1.5\cdot b}\\
\text{Risk weighted assets} &= \text{RWA} = K \cdot 12.5 \cdot EAD
\end{aligned}
$$

If the calculation in above results in a negative capital charge for any individual sovereign exposure, banks should apply a zero capital charge for that exposure.

In additiona, there are some modifications on the correlation for several categoriesï¼š
* Financial institutions meeting the following criteria:
    * Regulated Financial Institution swhose total assets are greater than or equal to USD 100 billion
    * Unregulated financial institutions, regardless of size.
    * The correlation formula is as follows:
$$
\begin{aligned}
\text{Correlation for FI} &= R_{FI} = 1.25 \cdot \left[0.12 \cdot \frac{1 - \exp^{-50 \cdot PD}}{1 - \exp^{-50}} + 0.24 \cdot \left( 1 - \frac{1 - \exp^{-50 \cdot PD}}{1 - \exp^{-50}} \right) \right] \\
\end{aligned}
$$
* Small or medium-sized entities (SMEs)
    * S is expressed as total annual sales in millions of euros with values of Sales falling in the range of equal to or less than â‚¬50 million or greater than or equal to â‚¬5 million.
    * The correlation formula is as follows:
$$
\begin{aligned}
\text{Correlation for SMEs} &= R_{SME} &= 0.12 \cdot \frac{1 - \exp^{-50 \cdot PD}}{1 - \exp^{-50}} + 0.24 \cdot \left( 1 - \frac{1 - \exp^{-50 \cdot PD}}{1 - \exp^{-50}} \right) - 0.04 \cdot \left(1 - \frac{S-5}{45} \right) \\
\end{aligned}
$$
* High volatility commercial real estate (HVCRE) sub-asset class of Special lending
    * The correlation formula is as follows:
$$ \text{Correlation for HVCREs} = R_{HVCRE} = 0.12 \cdot \frac{1 - \exp^{-50 \cdot PD}}{1 - \exp^{-50}} + 0.3 \cdot \left( 1 - \frac{1 - \exp^{-50 \cdot PD}}{1 - \exp^{-50}} \right)$$    

#### 3.4.2 Risk-weight functions for retail exposures
The capital Requirement and RWA formulas are the same for all the following exposures, which is:
$$
\begin{aligned}
\text{Capital Requirement} &= K &= \left[ N\left( \frac{1}{\sqrt{1-R}} \cdot G(PD) + \sqrt{\frac{R}{1-R}} \cdot G(0.999) \right) \cdot LGD - PD \cdot LGD \right]\\
\text{Risk weighted assets} &= \text{RWA} &= K \cdot 12.5 \cdot EAD
\end{aligned}
$$

The only difference is the correlation:
* Retail residential mortgage exposures:
$$\text{Correlation} = R = 0.15 $$

* Qualifying revolving retail exposures
$$\text{Correlation} = R = 0.04 $$

* Other retail exposures
$$ \text{Correlation} = R = 0.03 \cdot \frac{1 - \exp^{-35 \cdot PD}}{1 - \exp^{-35}} + 0.16 \cdot \left( 1 - \frac{1 - \exp^{-35 \cdot PD}}{1 - \exp^{-35}} \right)$$

#### 3.4.3 Explanatory note on the risk weights formula
* Unexpected Loss: Losses above expected levels
* Capital is set to ensure that unexpected losses will exceed this level of capital with only very low, fixed probability.
* Correlation:
    * Correlation is the weighted average of correlation for high PDs and correlation for low PDs for each asset class (except for Retail residential mortgage and QRRE):
        + Correlation for high PDs: 0.12 for corporate, bank and sovereign exposure, 0.03 for other retail exposure
        + Correlation for low PDs: 0.24 for corporate, bank and sovereign exposure, 0.16 for other retail exposure
        + Correlation k-factor: 50 for corporate, bank and sovereign exposure, 35 for other retail exposure since correlation from other retail asset class decreases at a slower paces.
    * Rationale for retail exposures have less correlation: retail customer tend to be more idiosyncratic ans less dependent on the economic ecycle than corporate exposures.
* Maturity: 
    * Interpretation 1: Both intuition and empirical evidence indicate that long-term credits are riskier than short-term credits and capital should increase with maturity.
    * Interpretation 2: Anticipation of additional capital requirements due to downgrades which is more likely in case of long-term credits.
* Maturity adjustment:
    * Maturity effects are stronger with low PDs than with high PDs. Intuitively, low PD borrowers have more 'potential' and more room for down-gradings than high PD borrowers.

```python
# Expected credit loss and unexpected credit loss explained
import numpy as np 
from scipy.stats import gamma
import matplotlib.pyplot as plt
a, a_stress = 6, 6
loc, loc_stress = 3, 6
scale, scale_stress = 1, 2

# Simulate loss 
mean, mean_stress = gamma.mean(a, loc, scale), gamma.mean(a_stress, loc_stress, scale_stress)
simu_loss, simu_loss_stress = np.linspace(gamma.ppf(0, a, loc, scale), gamma.ppf(0.9999, a, loc, scale), 10000), np.linspace(gamma.ppf(0, a_stress, loc_stress, scale_stress), gamma.ppf(0.9999, a_stress, loc_stress, scale_stress), 10000)
simu_freq, simu_freq_stress = gamma.pdf(simu_loss, a, loc, scale), gamma.pdf(simu_loss_stress, a_stress, loc_stress, scale_stress)

# Plotting
fig, ax = plt.subplots(figsize=(15, 6))
ax.plot(simu_loss, simu_freq, color='blue', label='Normal Condition')
ax.plot(simu_loss_stress, simu_freq_stress, color='red', label='Stress Condition')
ax.set_ylim(-0.04, np.max([simu_freq, simu_freq_stress]) + 0.01)
ax.set_xlim(np.min([simu_loss, simu_loss_stress])-0.01, np.max([simu_loss, simu_loss_stress])+0.01)
ax.set_xlabel('Potential Losses')
ax.set_ylabel('pdf')
# Expected loss annotation
ax.plot(np.repeat(mean, 2), [0, gamma.pdf(mean, a, loc, scale)], color='blue', linestyle='dashed')
ax.plot(np.repeat(mean_stress, 2), [0, gamma.pdf(mean_stress, a_stress, loc_stress, scale_stress)], color='red', linestyle='dashed')
ax.annotate("", xy=(gamma.ppf(0, a, loc, scale), 0), xycoords='data', xytext=(mean, 0), textcoords='data', arrowprops=dict(arrowstyle="<->"))
ax.text(4, -0.015, "Expected Loss(EL)")
# Unexpected loss annotation
ax.plot(np.repeat(gamma.ppf(0.975, a, loc, scale), 2), [0, gamma.pdf(gamma.ppf(0.975, a, loc, scale), a, loc, scale)], color='grey', linestyle='dashed')
ax.plot(np.repeat(gamma.ppf(0.975, a_stress, loc_stress, scale_stress), 2), [0, gamma.pdf(gamma.ppf(0.975, a_stress, loc_stress, scale_stress), a_stress, loc_stress, scale_stress)], color='grey', linestyle='dashed')
ax.annotate("", xy=(mean, 0), xycoords='data', xytext=(gamma.ppf(0.975, a, loc, scale), 0), textcoords='data', arrowprops=dict(arrowstyle="<->"))
ax.text(9.5, -0.015, "Unexpected Loss(UL)")
# Value at Risk
ax.annotate("", xy=(gamma.ppf(0, a, loc, scale), -0.025), xycoords='data', xytext=(gamma.ppf(0.975, a, loc, scale), -0.025), textcoords='data', arrowprops=dict(arrowstyle="<->"))
ax.text(7.5, -0.035, "Value at Risk")
# Confidence level
ax.fill(np.append(np.append(gamma.ppf(0.975, a, loc, scale), simu_loss[simu_loss >= gamma.ppf(0.975, a, loc, scale)]), simu_loss[len(simu_loss)-1]), 
        np.append(np.append(0, simu_freq[simu_loss >= gamma.ppf(0.975, a, loc, scale)]), 0), 'grey', alpha = 0.5)
ax.annotate("100% minus confidence level", xy=(15, 0.005), xycoords='data', xytext=(11, 0.12), textcoords='data', arrowprops=dict(arrowstyle="<-"))
ax.legend()
plt.show()
```

```python
# implement the RWA calculation
import numpy as np
from scipy.stats import norm
class RWAcredit:
    def __init__(self, assetclass, subclass, sl, fi, sales):
        self.assetclass = assetclass
        self.sl = sl
        self.fi = fi
        self.subclass = subclass
        self.sales = sales
        if self.assetclass.upper() in ['CORPORATE', 'SOVEREIGN', 'BANK']:
            self.corr_highPD = 0.12
            self.corr_lowPD = 0.24
            self.corr_expw = -50
            if self.fi == 1:
                self.corr_highPD *= 1.25
                self.corr_lowPD *= 1.25
            elif self.subclass.upper() == 'HVCRE':
                self.corr_lowPD = 0.3
        elif self.assetclass.upper() == 'RETAIL':
            self.corr_highPD = 0.03
            self.corr_lowPD = 0.16
            self.corr_expw = -35
                
    def calcCorr(self, pd):
        self.correlation = self.corr_highPD * (1-np.exp(self.corr_expw * pd)) / (1-np.exp(self.corr_expw)) + self.corr_lowPD * (1- (1-np.exp(self.corr_expw * pd)) / (1-np.exp(self.corr_expw)))
        if (self.assetclass.upper() in ['CORPORATE', 'SOVEREIGN', 'BANK']) & (self.subclass.upper() == 'SME') & (self.sales <= 50) & (self.sales >= 5):
            self.correlation -= 0.04 * (1- (self.sales - 5)/45)
        elif self.assetclass.upper() == 'RETAIL':
            if self.subclass.upper() == 'RESIDENTIAL MORTGAGE':
                self.correlation = 0.15
            elif self.subclass.upper() == 'QRRE':
                self.correlation = 0.04
                
    def calcMatadj(self, pd, m):
        if self.assetclass.upper() in ['CORPORATE', 'SOVEREIGN', 'BANK']:
            self.maturityB = np.square(0.11852 - 0.05478 * np.log(pd))
            self.maturityAdj = (1 + (m - 2.5)*self.maturityB) / (1 - 1.5 * self.maturityB)
        elif self.assetclass.upper() == 'RETAIL':
            self.maturityB = np.repeat(np.nan, len(pd))
            self.maturityAdj = np.repeat(1, len(pd))
    
    def calcRWA(self, pd, lgd, ead, m): 
        self.calcCorr(pd)
        self.calcMatadj(pd, m)
        self.extPD = norm.cdf(np.sqrt(1/(1-self.correlation)) * norm.ppf(pd, loc=0, scale=1) + np.sqrt(self.correlation/(1-self.correlation)) * norm.ppf(0.999, loc=0, scale=1), loc=0,scale=1)
        self.ulpct = (self.extPD - pd) * lgd
        self.K = self.ulpct * self.maturityAdj
        self.RWA = self.K * ead * 12.5       
```

```python
import matplotlib.pyplot as plt
# Define pd array
pd = np.linspace(0.0001, 0.1, num=10000)
# Corporate asset class
corp = RWAcredit(assetclass='CORPORATE', subclass='SME', sl='', fi=0, sales=100) 
corp.calcRWA(pd, lgd=0.45, ead=1, m=2.5)
# Retail asset class
retail = RWAcredit(assetclass='RETAIL', subclass='', sl='', fi=0, sales=np.nan) 
retail.calcRWA(pd, lgd=0.45, ead=1, m=np.nan)

# Plot the pd vs.RWA
plt.figure(figsize=(20, 5))
plt.suptitle('LGD = 0.45, EAD = 1, Maturity = 2.5')
# first plot - Correlation
ax1 = plt.subplot(131)
plt.plot(pd, corp.correlation, label='Corporate Asset Class')
plt.plot(pd, retail.correlation, label='Retail Asset Class')
ax1.set_title('PD vs. Asset Correlation')
ax1.set_xlabel('pd')
ax1.set_ylabel('Correlation')
ax1.legend()
# second plot - maturiy adjustment
ax2 = plt.subplot(132)
plt.plot(pd, corp.maturityAdj, label='Corporate Asset Class')
plt.plot(pd, retail.maturityAdj, label='Retail Asset Class')
ax2.set_title('PD vs. Maturity Adjustment')
ax2.set_xlabel('pd')
ax2.set_ylabel('Maturity Adjustment')
ax2.legend()
# third plot - RWA
ax3 = plt.subplot(133)
plt.plot(pd, corp.RWA, label='Corporate Asset Class')
plt.plot(pd, retail.RWA, label='Retail Asset Class')
ax3.set_title('PD vs. RWA')
ax3.set_xlabel('pd')
ax3.set_ylabel('RWA')
ax3.legend()
plt.show()
```

#### 3.5.2 PD
* Requirements for **corporate, sovereign and bank exposures**:  
    1. Corporate and Bank exposures: $ PD_i = \max{\left( \text{one-year} \, PD_i, 0.03\% \right)}$, where $i$ is the internal borrower grade to which that exposure is assigned.  
    2. Sovereign exposures: $PD_i = \text{one-year} \, PD_i$, where $i$ is the internal borrower grade to which that exposure is assigned.
    3. Default grade PD is 100%.
    4. Requirements for PD estimation:
        + a. Banks must use information and techniques that take appropriate account of the long-run experience when estimating the average PD for each rating grade. e.g, bank may use **one or more** of the three specific techniques:
            + internal default experience
            + mapping external data
            + statistical default models
        + b. The length of the underlying historical observation period used for PD estimation must be $>= 5$ years for at least one source. If the available observation period spans a longer period for any source, and this data are relevant and material, this longer period must be used.
* Requirements for **Retail expsoures**:
    1.  $ PD_i = \max{\left( \text{one-year} \, PD_i, 0.03\% \right)}$, where $i$ is the internal borrower grade to which the pool of retail exposures is assigned.

## 4. Treatment of expected losses and recognition of allowances
### 4.1 Expected losses
* EL for exposures other than SL subject to the supervisory slotting criteria:
    1. For corporate, sovereign, bank and retail exposures both not in default and not treated as hedged exposures under the double default treatment: Bank must calculate $EL = PD \times LGD$
    2. For corporate, sovereign, bank and retail exposures in default: Bank must use  the best estimate of expected loss
    3. For corporate, sovereign, bank and retail exposures treated as hedged exposures under the double default treatment: $EL = 0$
* EL for SL exposures subject to the supervisory slotting criteria: $EL = 8\% \times \text{RWA}$  

### 4.2 Allowances
* Under IFRS9:
    1. Specific allowances: Stage 3 allowances and partial write-offs
    2. General allowances: Stage 1 and 2 allowances
